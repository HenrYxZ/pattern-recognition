```python
import glob
import numpy as np
vlad_path_list = glob.glob("../vlad128_2/*")
vlad_path_list
```

```python
vlad = None

total = len(vlad_path_list)
i = 0
for vlad_path in vlad_path_list:
    image_vlad = np.loadtxt(vlad_path, delimiter=",")  
    if vlad is None:
        vlad = image_vlad
    else:
        vlad = np.vstack((vlad,image_vlad))
    if i % 10 == 0:
        print(str(i) + "/" + str(total))
    i += 1
np.savetxt("vlad128.csv", vlad, delimiter=",")
```

```python
def get_file_list(path):
    with open(path) as f:
        lines = f.read().splitlines()
    return lines
```
